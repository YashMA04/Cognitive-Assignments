```python
data={
    "subjects": ['UCS301','UCS302','UCS303','UCS304','UCS305'],
    "scores": [80,90,94,89,88]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
sns.barplot(x="subjects",y="scores",data=df,palette="coolwarm")
plt.title("Scores", fontsize=14, fontweight="bold")
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.7)
```

```python
df=pd.read_csv('Company.csv')
```

```python
df.head()
```

```python
sns.plot(df['total_profit'])
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=df, marker='o', color='green')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
for i in df.columns:
    if(i!='total_profit' and i!='total_units'):
        sns.lineplot(x='month_number', y=i , data=df, marker='o')
```

```python
total_sales = df.drop('month_number', axis=1).sum()


plt.figure(figsize=(12, 7))
sns.barplot(x=total_sales.index, y=total_sales.values, palette='muted')
plt.title('Total Sales and Profit by Product')
plt.ylabel('Total Sales/Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
