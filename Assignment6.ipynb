```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Q1

```python
x=np.linspace(-10,10,100)
```

```python
x
```

```python
m=2
y=m*(x**2)
y
```

```python
y2=m*np.sin(x)
```

```python
y2
```

```python
plt.plot(x,y,'r--',label="mx^2")
plt.plot(x,y2,'b-',label="msinX")
plt.legend()
plt.grid()
plt.title("PLot")
```

Q2

```python
data={
    "subjects": ['UCS301','UCS302','UCS303','UCS304','UCS305'],
    "scores": [80,90,94,89,88]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
sns.barplot(x="subjects",y="scores",data=df,palette="coolwarm")
plt.title("Scores", fontsize=14, fontweight="bold")
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.7)
```
