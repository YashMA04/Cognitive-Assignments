```python
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from collections import Counter
nltk.download('punkt')
nltk.download('stopwords')
```

```python
"""
Question 1
Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports, technology, food, books, etc.). 
    1. Convert text to lowercase and remove punctua on using re. 
    2. Tokenize the text into words and sentences. 
    3. Split using split() and word_tokenize() and compare how Python split and NLTK’s word_tokenize() differ. 
    4. Remove stopwords (using NLTK's stopwords list). 
    5. Display word frequency distribu on (excluding stopwords).
"""
```

```python
text = """Cricket is a game that I truly enjoy watching and playing. 
Many people in my country love cricket, and it brings everyone together. 
I usually play cricket with my friends on weekends, and we have so much fun. 
Watching cricket matches on TV is also exciting, especially when it's a close game. 
My favorite part of cricket is batting, but I also like bowling. Cricket has taught me teamwork, patience, and discipline. 
I believe cricket is more than just a sport — it's a feeling that connects people."""

# 1. Lowercase and remove punctuation
text_clean = re.sub(r'[^\w\s]', '', text.lower())
print("Text after converting to lowercase and removing punctuation:\n", text_clean)
```

```python
# 2. Tokenization
word_tokens = word_tokenize(text_clean)
sent_tokens = sent_tokenize(text)
print("Words:\n", word_tokens)
print("Sentences:\n", sent_tokens)
```

```python
# 3. Split() vs word_tokenize()
split_words = text.split()
print("Split() Output:\n", split_words)
print("word_tokenize() Output:\n", word_tokens)
```

```python
# 4. Remove stopwords
stop_words = set(stopwords.words('english'))
words_wo_stopwords = [w for w in word_tokens if w not in stop_words]
print("Filtered Words:\n", words_wo_stopwords)
```

```python
# 5. Word Frequency Distribution
word_freq = Counter(words_wo_stopwords)
print("Word Frequencies:\n", word_freq)

from nltk.probability import FreqDist
fdist = FreqDist(words_wo_stopwords)
fdist.plot(20, title="Word Frequency (without Stopwords)")
```

```python
"""
Question 2:
Using the same paragraph from Q1: 
    1. Extract all words with only alphabets using re.findall() 
    2. Remove stop words using NLTK’s stopword list 
    3. Perform stemming with PorterStemmer 
    4. Perform lemma za on with WordNetLemma zer 
    5. Compare the stemmed and lemma zed outputs and explain when you’d prefer one over the other. 
"""
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

# 1. Extract words with only alphabets
text_lower = text.lower()
alpha_words = re.findall(r'\b[a-zA-Z]+\b', text_lower)
print("Alphabet only Words:\n", alpha_words)
```

```python
# 2. Remove stopwords
filtered = [w for w in alpha_words if w not in stop_words]
```

```python
# 3. Stemming
stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in filtered]
print("Stemmed Words:\n", stemmed)
```

```python
# 4. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in filtered]
print("Lemmatized Words:\n", lemmatized)
```

```python
# 5. Comparison
print("Comparison:")
for s, l in zip(stemmed, lemmatized):
    print(f"Stem: {s} | Lemma: {l}")
print("Stemming is faster but may distort meaning; Lemmatization is better for meaningful text processing.")
```

```python
"""
Question 3:
Choose 3 short texts of your own (e.g., different news headlines, product reviews). 
    1. Use CountVectorizer to generate the Bag of Words representa on. 
    2. Use TfidfVectorizer to compute TF-IDF scores. 
    3. Print and interpret the top 3 keywords from each text using TF-IDF. 
"""
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

texts = [
    "The phone battery life is excellent and lasts all day.",
    "Camera quality is amazing in low light conditions.",
    "The screen is very bright and clear in sunlight."
]
```

```python
# 1. CountVectorizer (BoW)
cv = CountVectorizer()
bow = cv.fit_transform(texts)
print("Bag of Words:\n", cv.get_feature_names_out())
print(bow.toarray())
```

```python
# 2. TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(texts)
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python
# 3. Top 3 keywords per text
import numpy as np

feature_names = tfidf.get_feature_names_out()
for i, text in enumerate(texts):
    print(f"\nTop words in text {i+1}:")
    tfidf_scores = tfidf_matrix[i].toarray()[0]
    top_indices = np.argsort(tfidf_scores)[-3:][::-1]
    for idx in top_indices:
        print(f"{feature_names[idx]}: {tfidf_scores[idx]:.4f}")
```

```python
"""
Write 2 short texts (4–6 lines each) describing two different technologies (e.g., AI vs Blockchain). 
    1. Preprocess and tokenize both texts. 
        2. Calculate: 
            a. Jaccard Similarity using sets 
            b. Cosine Similarity using TfidfVectorizer + cosine_similarity() 
            c. Analyze which similarity metric gives be er insights in your case. 
"""
```

```python
from sklearn.metrics.pairwise import cosine_similarity

text1 = """Artificial Intelligence (AI) is changing the way we interact with technology. 
            AI systems are used in voice assistants, self-driving cars, and smart home devices. 
            Machine learning, a branch of AI, helps in making better decisions using data."""
text2 = """Machine Learning is a subset of Artificial Intelligence. 
            It allows systems to learn from data and improve their performance without being explicitly programmed. 
            Many AI applications like recommendation engines and image recognition use machine learning."""
```

```python
# 1. Preprocess
t1_tokens = set(word_tokenize(re.sub(r'[^\w\s]', '', text1.lower())))
t2_tokens = set(word_tokenize(re.sub(r'[^\w\s]', '', text2.lower())))
```

```python
# 2a. Jaccard Similarity
intersection = t1_tokens.intersection(t2_tokens)
union = t1_tokens.union(t2_tokens)
jaccard_sim = len(intersection) / len(union)
print("\nJaccard Similarity:", jaccard_sim)
```

```python
# 2b. Cosine Similarity
tfidf_sim = TfidfVectorizer().fit_transform([text1, text2])
cos_sim = cosine_similarity(tfidf_sim[0:1], tfidf_sim[1:2])
print("Cosine Similarity:", cos_sim[0][0])
```

```python
# 2c. Comparison
print("Cosine Similarity captures word importance better; Jaccard is simpler but may miss context.")
```

```python
"""
Question 5:
 Write a short review for a product or service. 
     1. Use TextBlob or VADER to find polarity & subjec vity for each review. 
     2. Classify reviews into Posi ve / Nega ve / Neutral. 
     3. Create a word cloud using the wordcloud library for all posi ve reviews. 
"""
```

```python
!pip install textblob
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Review
reviews = [
    "This laptop is incredibly fast and efficient!",
    "The customer service was terrible and unhelpful.",
    "It's okay, nothing exceptional but not bad either.",
    "The camera quality is fantastic and works great even in low light. Totally worth the price!",
    "The delivery was delayed and the packaging was damaged. I'm very disappointed with the service.",
    "The phone is okay. It works fine for basic tasks, but there's nothing special about it.",
    "I loved the user interface — it's smooth, modern, and very easy to navigate. Great job!",
    "Battery drains too quickly and the phone heats up a lot. Not happy with the purchase.",
    "Customer support was super helpful and solved my issue in just 10 minutes!"
]
```

```python
# 1. Sentiment using TextBlob and VADER
sid = SentimentIntensityAnalyzer()
for r in reviews:
    tb = TextBlob(r)
    vader = sid.polarity_scores(r)
    print(f"\nReview: {r}")
    print("TextBlob:", tb.sentiment)
    print("VADER:", vader)
```

```python
# 2. Classification
def classify(score):
    if score > 0.1: return "Positive"
    elif score < -0.1: return "Negative"
    else: return "Neutral"

labels = [classify(sid.polarity_scores(r)['compound']) for r in reviews]
print("\nSentiment Labels:", labels)
```

```python
# 3. WordCloud for positive reviews
positive_text = ' '.join([r for i, r in enumerate(reviews) if labels[i] == "Positive"])
wc = WordCloud(background_color='white').generate(positive_text)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
