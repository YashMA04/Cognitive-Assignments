```python
import random as rand
```

```python
my_list=[]
for i in range(100):
  my_list.append(rand.randint(100,900))
```

```python
my_list
```

```python
#odd numbers
count=0
for i in my_list:
  if i%2==1:
    print(i)
    count=count+1
```

```python
#even number
count=0
for i in my_list:
  if i%2==0:
    print(i)
    count=count+1
print("count is ",count)
```

```python
from sympy import *
count=0
for i in my_list:
  if isprime(i):
    print(i)
    count=count+1
print("count is ",count)
```
