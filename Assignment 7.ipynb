```python
import numpy as np
import random as rand
import pandas as pd
```

```python
seed=np.random.seed(102317052)
seed
```

```python
sales=np.random.randint(1000,5000+1,size=(12,4))
```

```python
cats=['Electronics','Clothing','Home','Sports']
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
df=pd.DataFrame(sales,columns=cats,index=months)
```

```python
df
```

```python
df.head()
```

```python
df.describe()
```

```python
total_per_cat=[]
for j in cats:
    cost=0
    for i in months:
        cost+=df[j][i]
    total_per_cat.append(cost)
        
```

```python
total_per_cat
```

```python
total_per_month = df.sum(axis=1)
```

```python
df['Total Sales'] = df.sum(axis=1)

total_per_month
```

```python
df['Growth Rate'] = df['Total Sales'].pct_change() * 100
df.head()
```

```python
df['Electronics'] = df['Electronics'] * 0.9  # 10% discount
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
for category in ['Electronics', 'Clothing', 'Home', 'Sports']:
    plt.plot(df.index, df[category], marker='o', label=category)

plt.title('Monthly Sales Trends per Category')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Electronics', 'Clothing', 'Home', 'Sports']])
plt.title('Sales Distribution per Category')
plt.xlabel('Product Category')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()
```
